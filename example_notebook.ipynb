## Explore Database Schema

```python
# View database schema
schema = agent.get_schema_info()

for table_name, columns in schema.items():
    print(f"\nðŸ“‹ Table: {table_name}")
    print("Columns:")
    for col in columns:
        nullable = "NULL" if col['nullable'] else "NOT NULL"
        print(f"  - {col['name']:20s} {col['type']:15s} {nullable}")
```

## Example Queries

### Query 1: Top Products by Revenue

```python
result = agent.query("What are the top 5 products by revenue?", return_visualization=True)

if result['success']:
    print(f"ðŸ” Generated SQL:")
    print(f"   {result['sql_query']}\n")
    
    print(f"ðŸ“Š Results:")
    display(result['results'])
    
    print(f"\nðŸ’¡ Interpretation:")
    print(f"   {result['interpretation']}")
    
    # Show visualization if available
    if result['visualization']:
        result['visualization'].show()
else:
    print(f"âŒ Error: {result['error']}")
```

## View Conversation History

```python
history = agent.get_conversation_history()

print(f"ðŸ’¬ Conversation History ({len(history)} messages):\n")

for i, msg in enumerate(history, 1):
    role = msg['role'].upper()
    content = msg['content'][:100] + "..." if len(msg['content']) > 100 else msg['content']
    print(f"{i}. {role}: {content}")
```

## Performance Metrics
